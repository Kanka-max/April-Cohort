```python
import pandas as pd

all_data = pd.read_csv("https://raw.githubusercontent.com/Kanka-max/ADS_REV_code/main/08%20Lesson%20Sales%20Analysis/all_data.csv")
```

```python
all_data.head()
```

```python
# create a copy of the dataframe

df  = all_data.copy()
df.head(2)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# Cleaning Data

# filling in missing values
# removing outliers
# removing duplicates
# aligning the features with their correct data type
```

```python
# fill in with 0
# fill in with the preceding value
# imputation - median, mean
# drop the missing values



```

```python
# drop the rows with missing data

df = df.dropna(how = "all")
df

```

```python
df.isna().sum()
```

```python
# "Or" Date column

temp_df = df[df["Order Date"].str[0:2] == "Or"]

temp_df
```

```python
df = df[df["Order Date"].str[0:2] != "Or"]

df
```

```python
# Quantity
# Price Each

df["Quantity Ordered"] = pd.to_numeric(df["Quantity Ordered"])
df["Price Each"] = pd.to_numeric(df["Price Each"])
```

```python
df.dtypes
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Day"] = pd.DatetimeIndex(df["Order Date"]).day
df["Month"] = pd.DatetimeIndex(df["Order Date"]).month
df["Year"] = pd.DatetimeIndex(df["Order Date"]).year
df["Hour"] = pd.DatetimeIndex(df["Order Date"]).hour
df["Minute"] = pd.DatetimeIndex(df["Order Date"]).minute
```

```python
df.info()
```

```python
df["Sales"] = df["Quantity Ordered"] * df["Price Each"]
```

```python
import matplotlib.pyplot as plt
```

```python
df.shape
```

```python
## What was the best month for sales? How much was earned that month?

# get the total sales

df["Sales"].sum()
```

```python
# create a subset
# to obtain the best month for sales

best_month_for_sales = df.groupby("Month")["Sales"].sum().reset_index()
best_month_for_sales
```

```python
# plot the data
plt.figure(figsize=(12,12))
plt.bar(best_month_for_sales["Month"], best_month_for_sales["Sales"])

#define the layout
plt.title("Best Month For Sales")
plt.xlabel("Month")
plt.ylabel("Sales in USD")

#render the chart - display
plt.show()
```

```python
## what city sold the most product

# city, sales

df["Purchase Address"]
```

```python
# split() - delimiter (tab, comma)
# strip() - takes care of the white spaces

## obtain the city
def get_city(address):
  return address.split(",")[1].strip(" ")


## obtain the state
def get_state(address):
  return address.split(",")[2].split(" ")[1]

## apply the functions
## apply lambda functions- one expression and one condition

df["City"] = df["Purchase Address"].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
df.head(2)
```

```python
city_sales = df.groupby("City")["Sales"].sum().reset_index()
city_sales.head(5)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(city_sales["City"], city_sales["Sales"])
plt.title("City Sales")
plt.xlabel("City")
plt.ylabel("Sales")
plt.xticks(rotation = "vertical")
plt.show()
```

```python
## what time should we display ads to maximize the likelihood of customers buying product
df["Count"] = 1

keys = [pair for pair, df in df.groupby(["Hour"])]
plt.figure(figsize=(10,10))
plt.plot(keys, df.groupby(["Hour"]).count()["Count"])
plt.xticks(keys)
plt.grid()
plt.show()
```
